```
import numpy as np
from numpy import savez_compressed
import os
import gdown
import gc
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
```

```
# Download correctly masked faces and store it in a zip file.
# If needs to download it manually please use the link: https://drive.google.com/file/d/1upkkdJ8x4QKivBqmhIrsAeLf2sR2A2_v/view?usp=sharing
correct_link = 'https://drive.google.com/uc?id=1upkkdJ8x4QKivBqmhIrsAeLf2sR2A2_v'
output_filename="correct.zip"
gdown.download(correct_link,output_filename,quiet=False)
```

```
# Unzip correctly masked faces obtained from the previous step
!mkdir Correct
!unzip "/content/correct.zip" -d "/content/Correct/"
```

```
# Get correctly masked data's folder names because information is stored in several folders
folder_path="/content/Correct"
folder_names=os.listdir(folder_path)
folder_names.sort()
print("There are total",len(folder_names),"folders in correcly masked data")
```

```
# Get correctly masked file names from above obtained folder names
exact_path=[]
for my_folder in folder_names:
  file_path=os.listdir(folder_path+"/"+my_folder)
  complete_path=[]
  for my_file in file_path:
    complete_path.append(folder_path+"/"+my_folder+"/"+my_file)
  exact_path.extend(complete_path)
exact_path=np.asarray(exact_path)
print("There are total",len(exact_path),"faces/files in correcly masked data")
```

```
# Store correctly wore masked faces in an array
correct_features=[]
for item in exact_path:
  img = load_img(item,color_mode="grayscale",target_size=(16, 16))
  img_array = img_to_array(img)
  correct_features.append(img_array)
correct_features=np.asarray(correct_features)
print("Shape of correcly masked array is:",correct_features.shape)
```

```
# Save correctly masked features. There is no need to store correctly masked labels as it doesn't have any subcategory.
savez_compressed('/content/drive/MyDrive/MSProjectData/Dataset1/Correct/correct_features.npz',  correct_features)
```

```
# Let's clear some memory using garbage collector
del correct_features
gc.collect()
```

```
# Download incorrectly masked faces and store it in a zip file.
# If needs to download it manually please use the link: https://drive.google.com/file/d/1h8hXuLYDIVupV3u80sd8CIdsTWDgExb3/view?usp=sharing
incorrect_link = 'https://drive.google.com/uc?id=1h8hXuLYDIVupV3u80sd8CIdsTWDgExb3'
output_filename="incorrect.zip"
gdown.download(incorrect_link,output_filename,quiet=False)
```

```
# Unzip incorrectly masked faces obtained from the previous step
!mkdir Incorrect
!unzip "/content/incorrect.zip" -d "/content/Incorrect/"
```

```
# Get incorrectly masked data's folder names because information is stored in several folders
folder_path="/content/Incorrect"
folder_names=os.listdir(folder_path)
folder_names.sort()
print("There are total",len(folder_names),"folders in incorrecly masked data")
```

```
# Get incorrectly masked file names from above obtained folder names
exact_path=[]
for my_folder in folder_names:
  file_path=os.listdir(folder_path+"/"+my_folder)
  complete_path=[]
  for my_file in file_path:
    complete_path.append(folder_path+"/"+my_folder+"/"+my_file)
  exact_path.extend(complete_path)
exact_path=np.asarray(exact_path)
print("There are total",len(exact_path),"faces/files in incorrecly masked data")
```

```
# Store incorrectly wore masked faces and labels in an array
incorrect_features=[]
incorrect_labels=[]
for item in exact_path:
  img = load_img(item,color_mode="grayscale",target_size=(16, 16))
  img_array = img_to_array(img)
  incorrect_features.append(img_array)
  incorrect_labels.append(item)
incorrect_features=np.asarray(incorrect_features)
incorrect_labels=np.asarray(incorrect_labels)
print("Shape of incorrecly masked array is:",incorrect_features.shape)
```

```
# Save incorrectly masked features and labels
from numpy import savez_compressed
savez_compressed('/content/drive/MyDrive/MSProjectData/Dataset1/Incorrect/incorrect_features.npz',  incorrect_features)
savez_compressed('/content/drive/MyDrive/MSProjectData/Dataset1/Incorrect/incorrect_labels.npz',  incorrect_labels)
```
